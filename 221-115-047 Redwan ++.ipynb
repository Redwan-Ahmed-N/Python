<br>
Create two DataFrames for two different departments, each with employee_id, name, and salary. Concatenate these DataFrames vertically to form a single DataFrame containing all employees.

**Instructions**:

2. Use the pd.concat() function to combine these DataFrames.
3. Concatenate the DataFrames by rows (axis=0) and reset the index to get a clean, continuous index.

```python
import pandas as pd
import numpy as np
Department_A = pd.DataFrame({'employee_id':[1,2,3],
                    'name': ['Redwan','Ahmed','Nahied '],
                    'salary': [50000, 60000, 55000]})
Department_B =pd.DataFrame ({'employee_id': [4,5,6],
                    'name': ['Archi', 'Arani', 'Basak'],
                    'salary': [50000, 60000, 55000]})
all_employees = pd.concat([Department_A, Department_B], axis=0)
all_employees
```

10: Create a visually appealing plot to represent the sine and cosine functions, with various adjustments and customizations, then save the figure to a file.

(i) Generate Data:
1. Use np.linspace to create an array x of 100 values between 0 and 10.
2. Compute the sine (np.sin(x)) and cosine (np.cos(x)) of x.


```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)
y_sin = np.sin(x)
y_cos = np.cos(x)
x,y_sin,y_cos
```

10(ii) Create a Figure with Subplots:

Create a figure with two subplots: one for a simple line plot and another for a scatter plot.


Line Plot Customizations:
1. In the first subplot, plot the sine function as a solid line and the cosine function as a dashed line.
2. Customize line colors (color='blue' for sine and color='orange' for cosine).
3. Set axes limits from 0 to 10 on the x-axis and -1.5 to 1.5 on the y-axis.
4. Add labels to the x-axis and y-axis, a title for the plot, and a legend for the sine and cosine curves.


```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0, 10, 100)
y_sine = np.sin(x)
y_cosine = np.cos(x)
fig, axs = plt.subplots(2, 1, figsize=(10, 8))
axs[0].plot(x, y_sine, color='blue', linewidth=2, label='Sine Function')
axs[0].plot(x, y_cosine, color='orange', linewidth=2, linestyle='--', label='Cosine Function')
axs[0].set_title('Sine and Cosine Functions', fontsize=16)
axs[0].set_xlabel('x', fontsize=14)
axs[0].set_ylabel('Function value', fontsize=14)
axs[0].set_xlim(0, 10)
axs[0].set_ylim(-1.5, 1.5)
axs[0].grid(color='gray', linestyle='--', linewidth=0.5)
axs[0].legend(fontsize=12)
axs[1].scatter(x, y_cosine, color='orange', label='Cosine Function', s=50)
axs[1].set_title('Scatter Plot of Cosine Function', fontsize=16)
axs[1].set_xlabel('x', fontsize=14)
axs[1].set_ylabel('cos(x)', fontsize=14)
axs[1].grid(color='gray', linestyle='--', linewidth=0.5)
axs[1].legend(fontsize=12)
plt.tight_layout()
plt.savefig('customized_line_scatter_plot.png', dpi=300, bbox_inches='tight')
plt.show()

```

Scatter Plot with Error Bars:
1. In the second subplot, plot a scatter plot of x vs. sin(x).
2. Add error bars to each data point using plt.errorbar. Use a random normal distribution to generate small errors (e.g., yerr=0.4).
3. Set a distinct color and shape (e.g., red squares) for the scatter plot.


```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)
y_sine = np.sin(x)
y_cosine = np.cos(x)

np.random.seed(0)
y_errors = np.random.normal(0.4, 0.1, size=y_sine.shape)

fig, axs = plt.subplots(2, 1, figsize=(10, 8))

axs[0].plot(x, y_sine, color='blue', linewidth=2, label='Sine Function')
axs[0].plot(x, y_cosine, color='orange', linewidth=2, linestyle='--', label='Cosine Function')
axs[0].set_title('Sine and Cosine Functions', fontsize=16)
axs[0].set_xlabel('x', fontsize=14)
axs[0].set_ylabel('Function value', fontsize=14)
axs[0].set_xlim(0, 10)
axs[0].set_ylim(-1.5, 1.5)
axs[0].grid(color='gray', linestyle='--', linewidth=0.5)
axs[0].legend(fontsize=12)

axs[1].errorbar(x, y_sine, yerr=y_errors, fmt='s', color='red', label='sin(x) with Error', markersize=5)
axs[1].set_title('Scatter Plot of sin(x) with Error Bars', fontsize=16)
axs[1].set_xlabel('x', fontsize=14)
axs[1].set_ylabel('sin(x)', fontsize=14)
axs[1].grid(color='gray', linestyle='--', linewidth=0.5)
axs[1].legend(fontsize=12)

plt.tight_layout()
plt.savefig('scatter_plot_with_error_bars.png', dpi=300, bbox_inches='tight')
plt.show()

```
